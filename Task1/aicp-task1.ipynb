```python
#TASK 1 : Calculate total price of choosen items 

item_codes = ["A1", "A2", "B1", "B2", "B3", "C1", "C2", "C3", "D1", "D2", "E1", "E2", "E3", "F1", "F2", "G1", "G2"]
item_descriptions = ["Case A1 Compact", "Case A2 Tower", "RAM 8 GB", "RAM 16 GB", "RAM 32 GB", "HDD 1 TB", "HDD 2 TB", "HDD 4 TB", "SSD 240 GB", "SSD 480 GB", "2nd HDD 1 TB", "2nd HDD 2 TB", "2nd HDD 4 TB", "Optical Drive DVD/Blu-Ray Player", "Optical Drive DVD/Blu-Ray Re-writer", "OS Standard Version", "OS Professional Version"]
item_prices = [75.00, 150.00, 79.99, 149.99, 299.99, 49.99, 89.99, 129.99, 59.99, 119.99, 49.99, 89.99, 129.99, 50.00, 100.00, 100.00, 175.00]

chosen_items = []
item_map = {code: description for code, description in zip(item_codes, item_descriptions)}

#display the available items
print("Available Items:")
for i in range(len(item_codes)):
    print(f"{item_codes[i]} - {item_descriptions[i]} - ${item_prices[i]:.2f}")

# choose items from list of items
case_choice = input("Choose one case (Enter item code): ")
chosen_items.append(case_choice)

ram_choice = input("Choose one RAM (Enter item code): ")
chosen_items.append(ram_choice)

hdd_choice = input("Choose one Main Hard Disk Drive (Enter item code): ")
chosen_items.append(hdd_choice)

# Calculate the total price
total_price = 200.00  
for item_code in chosen_items:
    total_price += item_prices[item_codes.index(item_code)]

# Output the chosen items and total price
print("\nChosen Items:")
for item_code in chosen_items:
    print(f"{item_map[item_code]} - ${item_prices[item_codes.index(item_code)]:.2f}")
print(f"Total Price: ${total_price:.2f}")


```

```python
#TASK 2 : Allow customer to buy additional items
item_codes = ["A1", "A2", "B1", "B2", "B3", "C1", "C2", "C3", "D1", "D2", "E1", "E2", "E3", "F1", "F2", "G1", "G2"]
item_descriptions = ["Case A1 Compact", "Case A2 Tower", "RAM 8 GB", "RAM 16 GB", "RAM 32 GB", "HDD 1 TB", "HDD 2 TB", "HDD 4 TB", "SSD 240 GB", "SSD 480 GB", "2nd HDD 1 TB", "2nd HDD 2 TB", "2nd HDD 4 TB", "Optical Drive DVD/Blu-Ray Player", "Optical Drive DVD/Blu-Ray Re-writer", "OS Standard Version", "OS Professional Version"]
item_prices = [75.00, 150.00, 79.99, 149.99, 299.99, 49.99, 89.99, 129.99, 59.99, 119.99, 49.99, 89.99, 129.99, 50.00, 100.00, 100.00, 175.00]

chosen_items = {}
total_price = 200.00

# Function to add an item to the chosen items
def add_item(item_code):
    index = item_codes.index(item_code)
    description = item_descriptions[index]
    price = item_prices[index]
    chosen_items[item_code] = (description, price)

# Function to calculate the total price
def calculate_total_price():
    global total_price
    total_price = 200.00  # Reset total price to basic set cost
    for item_code in chosen_items:
        total_price += chosen_items[item_code][1]

# Display available items and prices
print("Available Items:")
for i in range(len(item_codes)):
    print(f"{item_codes[i]} - {item_descriptions[i]} - ${item_prices[i]:.2f}")

while True:
    choice = input("Choose an item (Enter item code or 'done' to finish): ")
    
    if choice == "done":
        break
    
    if choice in item_codes:
        add_item(choice)
        print(f"{item_descriptions[item_codes.index(choice)]} added to your order.")
    else:
        print("Invalid item code. Please try again.")

calculate_total_price()

print("\nChosen Items:")
for item_code, (description, price) in chosen_items.items():
    print(f"{item_code} - {description} - ${price:.2f}")

print(f"\nTotal Price: ${total_price:.2f}")

```

```python
# TASK 3 : Apply the discount on additional items 
item_codes = ["A1", "A2", "B1", "B2", "B3", "C1", "C2", "C3", "D1", "D2", "E1", "E2", "E3", "F1", "F2", "G1", "G2"]
item_descriptions = ["Compact Case", "Tower Case", "8 GB RAM", "16 GB RAM", "32 GB RAM", "1 TB HDD", "2 TB HDD", "4 TB HDD", "240 GB SSD", "480 GB SSD", "1 TB HDD", "2 TB HDD", "4 TB HDD", "DVD/Blu-Ray Player", "DVD/Blu-Ray Re-writer", "Standard OS", "Professional OS"]
item_prices = [75.00, 150.00, 79.99, 149.99, 299.99, 49.99, 89.99, 129.99, 59.99, 119.99, 49.99, 89.99, 129.99, 50.00, 100.00, 100.00, 175.00]

chosen_items = []
chosen_additional_items = []
total_price = 200.00  

print("Choose one Case, one RAM, and one Main Hard Disk Drive:")
for i in range(3):
    item_code = input(f"Enter item code for {item_descriptions[i]}: ")
    chosen_items.append(item_code)
    index = item_codes.index(item_code)
    total_price += item_prices[index]

# Ask if the customer wants additional items
additional_items = input("Do you want to purchase additional items? (yes/no): ")
if additional_items.lower() == "yes":
    while True:
        item_code = input("Enter item code for additional item (or 'done' to finish): ")
        if item_code == "done":
            break
        elif item_code in item_codes:
            chosen_additional_items.append(item_code)
            index = item_codes.index(item_code)
            total_price += item_prices[index]
        else:
            print("Invalid item code. Please try again or enter 'done' to finish.")

if len(chosen_additional_items) == 1:
    # Apply a 5% discount
    discount = 0.05
    discount_description = "5% Discount"
elif len(chosen_additional_items) >= 2:
    # Apply a 10% discount
    discount = 0.10
    discount_description = "10% Discount"
else:
    discount = 0
    discount_description = "No Discount"
    
#discounted price calculation 
discounted_price = total_price - (total_price * discount)
money_saved = total_price - discounted_price


print("\nChosen Items:")
for item_code in chosen_items:
    index = item_codes.index(item_code)
    print(f"{item_descriptions[index]} - ${item_prices[index]:.2f}")
print(f"Total Price (excluding discount): ${total_price:.2f}")

if chosen_additional_items:
    print("\nAdditional Items:")
    for item_code in chosen_additional_items:
        index = item_codes.index(item_code)
        print(f"{item_descriptions[index]} - ${item_prices[index]:.2f}")
    print(f"\nDiscount Applied: {discount_description}")
    print(f"Money Saved: ${money_saved:.2f}")
    print(f"New Total Price (including discount): ${discounted_price:.2f}")
else:
    print("\nNo additional items chosen.")
```
